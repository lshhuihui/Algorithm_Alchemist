```python
import networkx as nx
import matplotlib.pyplot as plt

# 
G = nx.DiGraph()
G.add_edges_from([(0, 1), (0, 2), (1, 3), (1, 4), (2, 5), (2, 6)])

# 
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)
plt.show()

# 
pruned_edges = [(1, 4), (2, 6)]
G.remove_edges_from(pruned_edges)

# 
nx.draw(G, pos, with_labels=True, edge_color=['gray' if edge in pruned_edges else 'black' for edge in G.edges()])
plt.show()
```
