```python
'''
三大结构/六大容器

三大结构:
    顺序结构
    选择结构(分支) if else elif    and or not
    循环结构,for当知道起点的时候用, while不知道起点的时候选择用

    for循环语法:
        for i in range(起点,终点,步长): 
            循环体
    while循环语法:
        while 条件:
            循环体    
六大结构:
    字符串      str
    range序列   range
    列表        list
    元组        tuple
    字典        dict
    集合        set

'''
# x = eval(input("请输入一个整数:"))
x = 1
if x > 0 and x < 60:
    print("不合格")
elif x >= 60 and x < 70:
    print("及格")
else:
    print("优秀")

for i in range(1,3,1): # 左闭右开[1,101)
    if i % 2 == 0:
        print(i)
    print(i)

start = 1
end = 10
step = 1
while start < end:
    if start % 2 != 0:
        print(start)
    start += step # start = start + step 

# value x h h
# index 0 1 2
# 1.字符串str
s1 = "abcdefkasjalkdada" 
m1 = s1[1]
m2 = s1[2]
print(m1 + m2)
# 切片[起点:终点:步长]
print(s1[4:10:2])


s2 = 'hello world'
s3 = '''
hello world1
hello world2
'''
print(s1 == s2)
print(s3)
print(s1 == s3)
print(s2 == s3)

# range序列,range(1,5,2)是个range对象,range对象是不可变序列
print(range(1,5))
print(list(range(1,5,2)))
print(list(range(6))) # 6就是n,默认起点是0(default),步长是1(default)

# 列表list,可变序列,确实很好用,但效率很低
l1 = [2,6,100.9,"xhh",1213]
# values 2 6 4 1 1213
# index  0 1 2 3 4

# 元组tuple ()表示,元组不可变,不可修改,不可添加,坏处是比较死板,好处是效率高
l1 = ("北京","丽江","丽江","丽江") # 一段时间内不容易发生改变的数据,用元组
print(l1[2])

# 集合set  无序,不可重复,表逻辑
x = [1,2,2,3,3,4]
y = set(x)
x = list(y)
print(x)

# 字典dict = {"k1":v1,"k2":v2,....}
d =  {"北京":100,"上海":200,"广州":300}
print(d["上海"])

```
