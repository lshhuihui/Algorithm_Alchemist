```python
import networkx as nx
import matplotlib.pyplot as plt

# 创建图
G = nx.Graph()

# 添加带权重的节点
nodes = [
    ("Python基础", {"weight": 5}),
    ("数据类型", {"weight": 4}),
    ("函数", {"weight": 4}),
    ("模块导入", {"weight": 3})
]

G.add_nodes_from(nodes)

# 添加边
edges = [
    ("Python基础", "数据类型", {"weight": 0.8}),
    ("Python基础", "函数", {"weight": 0.8}),
    ("函数", "模块导入", {"weight": 0.6})
]

G.add_edges_from(edges)

# 计算节点布局
pos = nx.spring_layout(G, k=0.5)

# 绘制节点，字体大小与权重关联
nx.draw_networkx_nodes(G, pos, node_size=1000, node_color="lightblue")
nx.draw_networkx_edges(G, pos, width=2)

# 关键：根据权重设置字体大小
labels = {node: node for node in G.nodes()}
font_sizes = {node: G.nodes[node]["weight"] * 4 + 12 for node in G.nodes()}  # 权重映射到字体大小

nx.draw_networkx_labels(G, pos, labels=labels, font_size=font_sizes, font_weight="bold")

# 显示权重值
edge_labels = {(u, v): f"{d['weight']:.1f}" for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

plt.title("Python知识体系图")
plt.axis("off")
plt.show()
```
